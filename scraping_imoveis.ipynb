Importando bibliotecas

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import math
from time import sleep
```

Obtendo dados da p√°gina

```python
pagina = 1
```

```python
prefix = "https://www.imovelweb.com.br"
```

```python
webpage_response = requests.get(("https://www.imovelweb.com.br/comerciais-aluguel-sao-paulo-sp-pagina-{}.html").format(pagina))
```

```python
webpage = webpage_response.content
```

```python
soup = BeautifulSoup(webpage, "html.parser")
```

```python
total = math.ceil(int(soup.find("h1", {"class":"list-result-title"}).b.text.replace(".", ""))/20)
```

```python
realty_links = []
```

```python
for p in range(total):
    try:
        webpage_response = requests.get(("https://www.imovelweb.com.br/comerciais-aluguel-sao-paulo-sp-pagina-{}.html").format(pagina), verify = False)
        webpage = webpage_response.content
        soup = BeautifulSoup(webpage, "html.parser")
        realty_links.extend(soup.find_all("a", {"class": "go-to-posting"}))
        pagina += 1
        sleep(10)
    except:
        pass
```

```python
len(realty_links)
```

Incluindo links para im√≥veis em uma lista

```python
links = []
```

```python
for a in realty_links:
    links.append(prefix+a["href"])
```

```python
len(links)
```

```python
sites = pd.DataFrame(links)
```

```python
sites.to_csv('sites_aluguel1.csv', index=False, encoding='cp1252')
```
